### Total Number of activities are 448

```python
activities = pd.read_csv('../input/strava/activities.csv')
activities.shape
```

```python
activities.head(3)
```

```python
minimum_distance = min(activities['Distance'])
minimum_distance_ride = activities[activities['Distance'] == minimum_distance]
minimum_distance_ride_idx = minimum_distance_ride['Activity ID']
```

```python
import gpxpy
import gpxpy.gpx

gpx_file = open('../input/strava/activities/'+str(int(minimum_distance_ride_idx))+'.gpx', 'r')

gpx = gpxpy.parse(gpx_file)
 
lats, longs = [], []
for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            print('Point at ({0},{1}) -> {2}'.format(point.latitude, point.longitude, point.elevation))
            lats.append(point.latitude)
            longs.append(point.longitude)
```

### Display the shortest ride in the data

```python
import folium

points = []
for track in gpx.tracks:
    for segment in track.segments:        
        for point in segment.points:
            points.append(tuple([point.latitude, point.longitude]))
print(points)
ave_lat = sum(p[0] for p in points)/len(points)
ave_lon = sum(p[1] for p in points)/len(points)

# Load map centred on average coordinates
my_map = folium.Map(location=[ave_lat, ave_lon], zoom_start=14)

#add a markers
for each in points:  
    folium.Marker(each).add_to(my_map)

#fadd lines
folium.PolyLine(points, color="red", weight=2.5, opacity=1).add_to(my_map)
```
